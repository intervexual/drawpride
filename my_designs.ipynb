### 2024-12-19 Autism Spectrum Nautilus
Playing around with the design. [Posted to tumblr here](https://www.tumblr.com/intervex/770429034131750912/autism-spectrum-pride-flags-with-nautilus?source=share).

```python
d = draw.Drawing(500*2, 300*2)
draw_horiz_bars(d, ['#f7f7f7'])
spectrum_rainbow = ['#cf7280', '#dfa87c', '#eede77', '#95c77a', '#3bb07d', '#5bb9af', '#7bc2e0', '#97a7c0', '#b38da0']
draw_nautilus(d, spectrum_rainbow, orientation=HORIZONTAL, size_ratio=1.25, stretch_ratio=2.25)
filelocations = save_flag(d, 'plain_nautilus_on_whitish', outdir, same_folder=True, show_image=True)
```

```python
# Using the colours from the problematic ROYLG flag
d = draw.Drawing(500*2, 300*2)
draw_horiz_bars(d, ['#f7f7f7'])
draw_diagonal_stripes(d, ['#2cad74', '#98d570'] + ['#ffc65b']*6 + ['#f68249', '#e43546'])
draw_nautilus(d, spectrum_rainbow, orientation=HORIZONTAL, stretch_ratio=2.5, size_ratio=1.15)
filelocations = save_flag(d, 'diagonal_nautilus_on_roylg', outdir, same_folder=True, show_image=True)
```

```python
# riffing on https://www.tumblr.com/themogaidragon/762687565710213120/hi-i-just-wanted-to-ask-if-you-would-accept-an?source=share
# and https://www.tumblr.com/themogaidragon/701717045607317504/pathological-demand-avoidance?source=share
d = draw.Drawing(500*2, 300*2)
draw_horiz_bars(d, ['#2cad74', '#98d570', '#ffc65b', 'white', '#ffc65b', '#f68249', '#e43546'])
draw_circle(d, 'white', size_ratio=0.6)
draw_nautilus(d, spectrum_rainbow, orientation=HORIZONTAL, stretch_ratio=2.5, size_ratio=.5)
filelocations = save_flag(d, 'horizontal_nautilus_on_roylg', outdir, same_folder=True, show_image=True)
```
