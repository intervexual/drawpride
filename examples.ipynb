The following code needs to be run before any examples, and is setup that is shared between all of them:

## Basic stripes and piles
For our first example, here is a flag that is all vertical bars. To create a flag, you need to first create a canvas with draw.Drawing. You can then call a variety of functions which modify the canvas to (for example) add stripes, piles, rings, squares, etc.

```python
# example of vertical bars: the androgyne flag
d = draw.Drawing(500, 300)
draw_vert_bars(d, ['#FE007F', '#9832FF', '#00B8E7'])
filelocations = save_flag(d, 'androgyne', outdir, show_image=True)
```

## Rings and circles

### Bullseye pattern: Concentric Progress Flag
