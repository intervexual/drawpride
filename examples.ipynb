# Examples of use
The following code needs to be run before any examples, and is setup that is shared between all of them:

```python
import sys

sys.path.insert(0, 'drawflags/')
from pride_shapes import *
sys.path.insert(0, 'processflags/')
from utils import *
from IPython.display import Image
outdir = 'output/examples/'

# some useful colours
intersex_yellow = '#FDD70A'
intersex_purple = '#7A01AA'
trans_pink = '#F5A9B8'
trans_blue = '#5BCEFA'
trans_stripes = [trans_pink, trans_blue, 'white', trans_blue, trans_pink]
disability_stripes = ['#CF7280', '#EEDE77', '#E8E8E8', '#7BC2E0', '#3CB07D']
disability_grey = '#595959'
rainbow_stripes = ['#E40303', '#FF8C00', '#FFED00', '#008026', '#24408E', '#732982']
```

## Basic stripes and piles
For our first example, here is a flag that is all vertical bars. To create a flag, you need to first create a canvas with draw.Drawing. You can then call a variety of functions which modify the canvas to (for example) add stripes, piles, rings, squares, etc.

```python
# example of vertical bars: the androgyne flag
d = draw.Drawing(500, 300)
draw_vert_bars(d, ['#FE007F', '#9832FF', '#00B8E7'])
filelocations = save_flag(d, 'androgyne', outdir, show_image=True)
```

### Horizontal stripes and a pile: Demisexual flag

```python
# example of horizontal bars AND a pile: the demisexual flag
d = draw.Drawing(500, 300)
draw_horiz_bars(d, ['white', 'white', '#6e0170', '#d2d2d2', '#d2d2d2'])
draw_pile(d, 'black', size_ratio=1.1)
filelocations = save_flag(d, 'demisexual', outdir, show_image=True)
```

### Diagonal stripes: Disability pride

```python
# example of diagonal stripes: diasbility pride
d = draw.Drawing(500, 300)
# the colours are defined in examplesupport.py
draw_diagonal_stripes(d, [disability_grey]*5 + disability_stripes + [disability_grey]*5)
filelocations = save_flag(d, 'disability', outdir, show_image=True)
```

### Concentric Bezier Curves: the Mental Health Flag

```python
# example of concentric beziers: the mental health flag
d = draw.Drawing(500, 300)
draw_concentric_beziers(d, ['#1d351d', '#3c7737', '#5eb039', '#efd428'])
filelocations = save_flag(d, 'mentalhealth', outdir, show_image=True)
```

## Rings, squares, and crosses
Functions for creating rings, squares, and crosses, which I have grouped together since they're common themes in intersex-related flags.
### A simple ring: the Intersex flag

```python
# example of a ring flag: the intersex flag
d = draw.Drawing(500, 300)
draw_horiz_bars(d, [intersex_yellow])
draw_transparent_ring(d, intersex_purple)
filelocations = save_flag(d, 'intersex', outdir, show_image=True)
```

### A segmented ring: the Allodisabled-Intersex Flag

```python
# example of a segmented ring flag: the allodisabled-intersex flag (intersex with unrelated disability)
d = draw.Drawing(500, 300)
draw_horiz_bars(d, [intersex_yellow])
draw_segmented_ring(d, disability_stripes[::-1], border_fill=disability_grey)
filelocations = save_flag(d, 'allodisabled_intersex', outdir, show_image=True)
```

### Concentric rings: the Co-disabled intersex flag

```python
# example of a concentric ring flag: the codisabled-intersex flag (intersex with a frequently correlated disability)
d = draw.Drawing(500, 300)
draw_horiz_bars(d, [disability_grey])
draw_concentric_rings(d, disability_stripes[::-1])
filelocations = save_flag(d, 'codisabled_intersex', outdir, show_image=True)
```

### Putting a flag inside a ring: the Trans-Intersex flag

```python
# example of an inset flag: intersex and transgender flag
d = draw.Drawing(500, 300)
draw_inset_into_intersex(d, trans_stripes, intersex_yellow, intersex_purple, orientation=HORIZONTAL)
filelocations = save_flag(d, 'trans_intersex', outdir, show_image=True)
```

### Putting a cross into a ring: The Ipsogender flag

### A square: the Endosex flag

### Multiple piles: the Progress Pride flag
A "pile" is the vexillology term for one of the triangles that goes on the side.

## Infinity symbols
Please note that solid white lemniscates have been used as a symbol of the Métis in their flags since the early 1800s. The Métis are one of the major Indigenous groups living in what is now Canada. There are many ways to style infinity symbols and, speaking as an autistic person, I think it is inappropriate for autistic flags to use solid white lemniscates.
### Métis style lemniscates: the Queer Métis flag

```python
# example of a flag with a Métis lemniscate: the queer Métis flag
d = draw.Drawing(500, 300)
draw_horiz_bars(d, rainbow_stripes)
draw_metis_lemniscate(d, 'white')
filelocations = save_flag(d, 'queer_metis', outdir, show_image=True)
```

#### Adding a border: the Intersex Métis flag
