
## 2024-12-20: Autistic Flags

Made some autistic flag designs.

### Autistic Pride (Day)
First one is a general autistic flag, inspired by the official Autistic Pride Day logo from Australia. Rather than use a gradient, I use solid colours from [this autistic pride flag](https://commons.wikimedia.org/wiki/File:The_Autistic_Pride_Flag_Without_Metis_Infinity.svg).

```python
# Making a flag inspired by the Autistic Pride Day (Australia) logo
d = draw.Drawing(500*2, 300*2)
draw_horiz_bars(d, ['white'])
draw_concentric_infinities(d, ['#2cad74', 'white', '#ffc65b', 'white', '#e43546'], 'white', size_ratio=0.7)
filelocations = save_flag(d, 'autistic_concentric', outdir, same_folder=True, show_image=True)
```

### Codisabled Autistic
A flag for autistic people who have commonly correlated disabilities such as ADHD, EDS, ARFID, APD, etc. Colours are from the disability pride flag.

```python
d = draw.Drawing(500*2, 300*2)
draw_horiz_bars(d, [disability_grey])
draw_concentric_infinities(d, disability_stripes, disability_grey, size_ratio=0.7)
filelocations = save_flag(d, 'autistic_codisabled', outdir, same_folder=True, show_image=True)
```

### Autistic and Mad

A flag for autistic people who are also Mad (in a Mad Pride) kind of way. Colours are from the mad pride flag plus the grey of the disability pride flag.

```python
# codisabled but specifically mad
d = draw.Drawing(500*2, 300*2)
draw_horiz_bars(d, [disability_grey])
draw_concentric_infinities(d, ['#796098', '#bf7998', '#796098', '#bf7998'], disability_grey, size_ratio=0.7)
filelocations = save_flag(d, 'mad_autistic', outdir,  same_folder=True, show_image=True)
```

### Autism As Disability
Another version, using the disability colours but a white background.

```python
d = draw.Drawing(500*2, 300*2)
draw_horiz_bars(d, ['white'])
draw_concentric_infinities(d, [disability_grey] + disability_stripes + [disability_grey], 'white', size_ratio=0.7)
filelocations = save_flag(d, 'autistic_as_disability', outdir, same_folder=True, show_image=True)
```

### Yet Another Variation
Just when I thought I was done playing with this idea, I got another, and I rather like it. This uses all five colours from the autistic pride flag that I consider inappropriate because of its usage of a Mtis style infinity symbol.
```python
d = draw.Drawing(500*2, 300*2)
draw_horiz_bars(d, [disability_grey])
draw_concentric_infinities(d, ['#2cad74', '#98d570', '#ffc65b',  '#f68249', '#e43546'], disability_grey, size_ratio=0.7)
filelocations = save_flag(d, 'autistic_hybrid', outdir, same_folder=True, show_image=True)
```
